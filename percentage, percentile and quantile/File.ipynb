# Percentile

The p-th percentile is the value below which p% of the observations fall.

Formula:

$$
P = \frac{p}{100} \times (n + 1)
$$

Where:

- $p$ = desired percentile (e.g., 25, 50, 90)  
- $n$ = total number of observations  
- $P$ = position of percentile in sorted data  

Example:

50th percentile = Median


# Quartiles

Quartiles divide sorted data into 4 equal parts.

- $Q_1$ = First Quartile (25th percentile)  
- $Q_2$ = Second Quartile (Median, 50th percentile)  
- $Q_3$ = Third Quartile (75th percentile)  

Interquartile Range (IQR):

$$
IQR = Q_3 - Q_1
$$

IQR measures the spread of the middle 50% of data.
