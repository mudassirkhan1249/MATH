# CLT

```python
population = np.random.randint(0,100,10000)
df = pd.DataFrame({"Population" : population})
```

```python
plt.figure(figsize = (4,3))
sns.kdeplot(df['Population'])
```

```python
sample_means = []
for i in range(50):
    sample_data = []
    for j in range(500):
        sample_data.append(np.random.choice(population))
    sample_means.append(np.mean(sample_data))
```

```python
df = pd.DataFrame({'sample' : sample_means})
```

```python
plt.figure(figsize = (4,3))
sns.kdeplot(df['sample'])
```

```python
plt.figure(figsize = (4,3))
sns.histplot(df['sample'], kde = True)
```
