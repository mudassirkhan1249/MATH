# Skewness

Skewness measures the asymmetry of a distribution around its mean.

If skewness = 0 → data is perfectly symmetric  
If skewness > 0 → positively skewed (right tail longer)  
If skewness < 0 → negatively skewed (left tail longer)

Population Formula:

$$
\text{Skewness} = 
\frac{\frac{1}{N}\sum_{i=1}^{N}(x_i - \mu)^3}
{\left(\sqrt{\frac{1}{N}\sum_{i=1}^{N}(x_i - \mu)^2}\right)^3}
$$

Where:

- $x_i$ = each data value  
- $\mu$ = population mean  
- $N$ = total number of observations  

---

Sample Skewness:

$$
g_1 =
\frac{\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^3}
{\left(\sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2}\right)^3}
$$

Where:

- $\bar{x}$ = sample mean  
- $n$ = sample size  

---

Interpretation:

- Positive Skew → Mean > Median  
- Negative Skew → Mean < Median  
- Near 0 → Approximately symmetric distribution
