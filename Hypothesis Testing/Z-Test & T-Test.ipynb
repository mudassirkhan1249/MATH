## Hypothesis Testing (Z-Test) – Example

### Scenario
Imagine you work for an e-commerce company, and your team is responsible for analyzing customer purchase data. You want to determine whether a new website design has led to a significant increase in the average purchase amount compared to the old design.

### Data
You have collected data from a random sample of 30 customers who made purchases on the **old website design** and 30 customers who made purchases on the **new website design**.

You have the purchase amounts for both groups.

#### Old Website Design Data
[45.2, 42.8, 38.9, 43.5, 41.0, 44.6, 40.5, 42.7, 39.8, 41.4,  
44.3, 39.7, 42.2, 41.5, 39.6, 44.0, 43.1, 38.7, 43.9, 42.0,  
41.9, 42.8, 43.7, 41.3, 40.9, 42.5, 41.6, 42.5, 43.1, 44.0]

#### New Website Design Data
[48.5, 49.1, 50.2, 47.8, 48.7, 49.9, 48.0, 50.5, 49.8, 49.6,  
50.1, 48.6, 48.3, 49.0, 50.0, 48.4, 49.3, 49.5, 48.8, 50.6,  
50.4, 48.1, 49.2, 50.7, 51.3, 46.9, 47.8, 48.0, 49.4, 50.2]

#### Population Standard Deviation
σ = 2.5

---

```python
old_data = np.array([45.2, 42.8, 38.9, 43.5, 41.0, 44.6, 40.5, 42.7, 39.8, 41.4, 44.3, 39.7, 42.2, 41.5, 39.6, 44.0, 43.1, 38.7, 43.9, 42.0, 41.9, 42.8, 43.7, 41.3, 40.9, 42.5, 41.6, 42.5, 43.1, 44.0])
new_data = np.array([48.5, 49.1, 50.2, 47.8, 48.7, 49.9, 48.0, 50.5, 49.8, 49.6, 50.1, 48.6, 48.3, 49.0, 50.0, 48.4, 49.3, 49.5, 48.8, 50.6, 50.4, 48.1, 49.2, 50.7, 51.3, 46.9, 47.8, 48.0, 49.4, 50.2])
```

```python
mean_new = np.mean(new_data)
mean_old = np.mean(old_data)
pop_std = 2.5
no_sample = len(new_data)
alpha = 0.05
```

```python
z_cal = (mean_new - mean_old) / (pop_std / np.sqrt(no_sample))
```

```python
z_table = st.norm.ppf(alpha)
```

```python
if z_table < z_cal :
    print("Ha is right")
else :
    print("Ho is a right")
```

# T-Test

#### `Q1. A manufacturer claims that the average weight of a bag of potato chips is 150 grams. A sample of 25 bags is taken, and the average weight is found to be 148 grams, with a standard deviation of 5 grams. Test the manufacturer’s claim using a one-tailed t-test with a significance level of 0.05.`

```python
p_mean = 150
s_mean = 148
n_samples = 25
std = 5
alpha = 0.05
```

```python
t_table = st.t.ppf(alpha, n_samples - 1)
t_table
```

```python
t_cal = (s_mean - p_mean) / (std / np.sqrt(n_samples))
t_cal
```

```python
if z_table < z_cal :
    print("Ha is right")
else :
    print("Ho is a right")
```

#### `Q2. A company wants to test whether there is a difference in productivity between two teams. They randomly select 20 employees from each team and record their productivity scores. The mean productivity score for Team A is 80 with a standard deviation of 5, while the mean productivity score for Team B is 75 with a standard deviation of 6. Test at a 5% level of significance whether there is a difference in productivity between the two teams.`

```python
team_A = 80
team_B = 75
std_A = 5
std_B = 6
n = 20
alpha = 0.025
```

```python
t_table = st.t.ppf(1 - alpha, 20+20-2)
t_table
```

```python
t_cal = (team_A - team_B) / np.sqrt((std_A**2/n) + (std_B**2/n))
t_cal
```

```python
if z_table < z_cal :
    print("Ha is right")
else :
    print("Ho is a right")
```

#### `Q3. A company wants to test whether a new training program improves the typing speed of its employees. The typing speed of 20 employees was recorded before and after the training program. The data is given below. Test at a 5% level of significance whether the training program has an effect on the typing speed of the employees.`

#### Data:
##### Before: 50, 60, 45, 65, 55, 70, 40, 75, 80, 65, 70, 60, 50, 55, 45, 75, 60, 50, 65, 70

##### After: 60, 70, 55, 75, 65, 80, 50, 85, 90, 70, 75, 65, 55, 60, 50, 80, 65, 55, 70, 75

```python
mean_B = np.mean(befr)
mean_A = np.mean(aftr)
n = 20
alpha = 0.05/2
std_B = np.std(befr)
std_A = np.std(aftr)

mean_B, mean_A, n, alpha, std_A, std_B
```

```python
t_table = st.t.ppf(1-alpha, 20-1)
t_table
```

```python
t_cal = (mean_A - mean_B) / np.sqrt((std_A**2/n) + (std_B**2/n))
t_cal
```

```python
if z_table < z_cal :
    print("Ha is right")
else :
    print("Ho is a right")
```
