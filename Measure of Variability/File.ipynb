# Range

The formula for Range is:

$$
\text{Range} = \text{Maximum Value} - \text{Minimum Value}
$$

Or mathematically:

$$
R = X_{\text{max}} - X_{\text{min}}
$$

Where:

- $X_{\text{max}}$ = highest value in the dataset  
- $X_{\text{min}}$ = lowest value in the dataset  
- $R$ = range

# Mean Absolute Deviation (MAD)

The formula for Mean Absolute Deviation is:

$$
MAD = \frac{1}{n} \sum_{i=1}^{n} |x_i - \bar{x}|
$$

Where:

- $n$ = total number of observations  
- $x_i$ = each data value  
- $\bar{x}$ = mean of the data  
- $|x_i - \bar{x}|$ = absolute deviation from the mean


## By Graphical Way

```python
secA = np.array([45, 48, 52, 50, 47, 53, 49, 51, 46, 59])
secB = np.array([5, 10, 20, 30, 40, 60, 70, 80, 90, 95])
no = np.arange(1, 11)
m = secA.mean()

secA.mean(), secB.mean()
```

```python
plt.figure(figsize = (10, 3))

plt.scatter(secA, no, label = 'secA')
plt.scatter(secB, no, label = 'secB')
plt.plot([m, m, m, m, m, m, m, m, m, m], no, c = 'red', label = 'm')

plt.legend()
plt.show()
```

## By Mathematical Way

```python
MAD_secA = np.sum(np.abs(secA - m)) / len(secA)
MAD_secB = np.sum(np.abs(secB - m)) / len(secB)
```

```python
print(MAD_secA , "," , MAD_secB)
```

## Variance and Standard Deviation

### Variance

Variance measures how far the data points are spread out from the mean.

$$
\sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2
$$

Where:

- $x_i$ = each data value  
- $\mu$ = population mean  
- $N$ = total number of values  
- $\sigma^2$ = variance  

---

### Standard Deviation

Standard Deviation is the square root of variance.  
It shows the average spread of data from the mean.

$$
\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2}
$$

Where:

- $\sigma$ = standard deviation  
- $\sigma^2$ = variance  


```python
print(np.std(secA), ',', np.std(secB))
```

```python
print(np.var(secA), ',', np.var(secB))
```
